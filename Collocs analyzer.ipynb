```python
import json
import re
import pandas as pd
from tqdm.notebook import tqdm
```

```python
category = "superheroes"
```

```python
with open(category+"_attribute_collocs.json", 'r') as file:
    data = json.load(file)
```

```python
collocs_counts = []
```

```python
for k,v in data.items():
    collocs_counts.append({'key': k, 'count': len(v)})
```

```python
df = pd.DataFrame(collocs_counts)
```
